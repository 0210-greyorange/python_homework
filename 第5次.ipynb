## 实现精灵的类化

```python
class pokemon():
    def __init__(self, poke):
        self.s = poke[0]
        self.名称 = poke[1]
        self.种类1 = poke[2]
        self.种类2 = poke[3]
        self.体力值 = poke[4]
        self.攻击力 = poke[5]
        self.防御力 = poke[6]
        self.魔法攻击 = poke[7]
        self.魔法防御 = poke[8]
        self.速度 = poke[9]
        self.代数 = poke[10]
        self.传奇 = poke[11]
        self.五维总值 = poke[5]+poke[6]+poke[7]+poke[8]+poke[9]
        self.攻防差值 = poke[5]-poke[6]
        self.属性名称 = ['s', '名称', '种类1', '种类2', '体力值', '攻击力', '防御力', '魔法攻击', '魔法防御', '速度', '代数', '传奇', '五维总值', '攻防差值']
        self.属性值 = [i for i in poke] + [poke[5]+poke[6]+poke[7]+poke[8]+poke[9], poke[5]-poke[6]]
    def findByName(self, name):
        if name==self.名称:
            return True
    def findById(self, s):
        if s==self.s:
            return True
    def display(self):
        for i in range(len(self.属性名称)):
            print(self.属性名称[i] + "：" + str(self.属性值[i]))
        print('\n')
```

## 导入文件并实现精灵的实例化

```python
import csv
path = "C:\学习资料\大二下\python数据分析\作业\pokemon.csv"
f = open(path, 'r')
next(f)
reader = csv.reader(f)
infos = []
for info in reader:
    for i in (0,4,5,6,7,8,9,10):
        info[i] = int(info[i])
    infos.append(info)
f.close()
pokes = []
for i in infos:
    poke = pokemon(i)
    pokes.append(poke)
```

## 基于名称查询某个精灵的所有属性值

```python
for i in pokes:
    if i.findByName('Metapod')==True:
        i.display()
        break
```

## 基于名称查询某个（或某些）精灵的所有属性值

```python
for i in pokes:
    if i.findById(6)==True:
        i.display()
```

## 实现选择排序算法的类化

```python
class SelectSortByStrenth():
    def __init__(self, a, x):
        for i in range(len(a[0].属性名称)):
            if x==a[0].属性名称[i]:
                属性_index = i
                break
        for i in range(len(a)):
            max_index = i
            for j in range(i+1, len(a)):
                if a[j].属性值[属性_index] > a[max_index].属性值[属性_index]:
                    max_index = j
            a[i], a[max_index] = a[max_index], a[i]       
```

```python
def sort_and_print(a, x, n):
    SelectSortByStrenth(a, x)
    print('按' + x + '实现降序排序的前五个精灵的属性值为：\n')
    for i in range(n):
        a[i].display() 
```

## 按五维总值、攻防差值以及所有可选择的数值型属性对精灵进行降序排列

```python
sort_and_print(pokes, '五维总值', 5)
sort_and_print(pokes, '攻防差值', 5)
sort_and_print(pokes, '体力值', 5)
sort_and_print(pokes, '攻击力', 5)
sort_and_print(pokes, '防御力', 5)
sort_and_print(pokes, '魔法攻击', 5)
sort_and_print(pokes, '魔法防御', 5)
sort_and_print(pokes, '速度', 5)
```

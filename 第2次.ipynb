```python
txtFile = open(r'C:\学习资料\大二下\python数据分析\作业\小项目文字.txt', encoding = 'utf-8')

result = []
num = 0
for line in txtFile:
    for i in line.split("。"):
        if i != "\n":
            num += 1
            i += "。"
            dic = {"第" + str(num)+ "句" : i}
            for j in i:
                if j in dic:
                    dic[j] += 1
                else:
                    dic[j] = 1
            result.append(dic)
            
for dic in result:
    for key,value in dic.items():
        if type(value) == str :
            print(key + "：" + value)
        else:            
            print(key + "：" + str(value), end = "      ")
    print("\n" * 3)
```

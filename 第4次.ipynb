## 导入数据

```python
import csv
path = "C:\学习资料\大二下\python数据分析\作业\pokeman.csv"
f = open(path, 'r')
next(f)
reader = csv.reader(f)
infos = []
for info in reader:
    for i in (0,4,5,6,7,8,9,10):
        info[i] = int(info[i])
    infos.append(info)
f.close()
```

## 将五维总值和攻防差值加入精灵信息

```python
for info in infos:
    fiveadd = 0
    for i in (5,6,7,8,9):
        fiveadd += info[i]
    info.append(fiveadd)
    info.append(info[5]-info[6])
```

## 冒泡排序降序排列函数

```python
def bubbleDownSort(a, x):
    for i in range(len(a)):
        for j in range(len(a)-i-1):
            if a[j][x] < a[j+1][x]:
                a[j], a[j+1] = a[j+1], a[j]
    return a
```

## 输出五维总值最大的精灵信息

```python
infos = bubbleDownSort(infos, 12)
print(infos[0][0:12])
```

## 输出攻防差值最大的精灵信息

```python
infos = bubbleDownSort(infos, 13)
print(infos[0][0:12])
```

## 冒泡排序降序并且输出前五个精灵的信息

```python
def sort_and_print(a, x, n):
    a = bubbleDownSort(a, x)
    for i in range(5):
        print(a[i][0:12])
```

## 按攻击力对精灵进行排序，并用冒泡排序降序排列输出前五个精灵的信息

```python
sort_and_print(infos, 5, 5)
```

## 按魔法攻击力对精灵进行排序，并用冒泡排序降序排列输出前五个精灵的信息

```python
sort_and_print(infos, 7, 5)
```
